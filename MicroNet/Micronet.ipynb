```python
import tensorflow as tf
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, Input, Conv1D
from tensorflow.keras.models import Model
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import SGD
import random as r
```

```python
x = [0.1*i -2 for i in range(0,81)]
y = [1.6*max(0,s-2) for s in x]

x = np.asarray(x)
y = np.asarray(y)
```

```python
bias = tf.keras.initializers.RandomUniform(minval=0, maxval=1)

x_in = Input(shape=(1,))
initializer= tf.keras.initializers.RandomUniform(minval=0, maxval=1.5)
s = Dense(1, activation='relu', kernel_initializer=initializer)(x_in)

model = Model(inputs=x_in, outputs=s)
model.summary()
```

```python
sgd = tf.keras.optimizers.SGD(lr=0.05)

model.compile(optimizer=sgd, loss='mean_squared_error', metrics=['accuracy'])
```

```python
model.fit(x, y, epochs=10, batch_size=10)
```

```python
y_predicted = model.predict(x)

plt.plot(x,y)
plt.plot(x,y_predicted)
plt.show()
```

```python
# ========== Max values ===========
```

```python
def category(x): # 1 if first number is bigger, 2 if equal, 3 if smaller
    if x[0] > x[1]:
        return 0
    else:
        return 1
```

```python
x_train = [[r.randint(-5,5)/5.0, r.randint(-5,5)/5.0] for _ in range(0,201)]
y_train = [category(x) for x in x_train]
x_train = np.asarray(x_train)
y_train = np.asarray(y_train)
y_train = to_categorical(y_train)
x_train, y_train
```

```python
x_in = Input(shape=(2,))
initializer = tf.keras.initializers.RandomUniform(minval=0, maxval=1.5)
s = Dense(5, activation='relu', kernel_initializer=initializer)(x_in)
s = Dense(2, activation='softmax', kernel_initializer=initializer)(s)

model = Model(inputs=x_in, outputs=s)
model.summary()
```

```python
model.compile(optimizer='SGD', loss='categorical_crossentropy')
```

```python
model.fit(x_train,y_train, epochs=100, batch_size=10)
```

```python
x = np.linspace(-2,2,num=20)
y = 1 + 2/(1+x**2) + 1/(1+(x-3)**2)

plt.plot(x,y)
```

```python
x_in = Input(shape=(1,))
s = Dense(100, activation='relu')(x_in)
s = Dense(25, activation='relu')(x_in)
s = Dense(1, activation='relu')(s)

model = Model(inputs=x_in, outputs=s)
model.summary()
```

```python
optim = tf.keras.optimizers.SGD(lr=0.03)
model.compile(optimizer=optim, loss="mean_squared_error")
model.fit(x, y, epochs=250, batch_size=10)
```

```python
y_pred = model.predict(x)

plt.plot(x,y)
plt.plot(x,y_pred)
```

```python
loss = model.history.history['loss']
plt.plot(loss)
plt.show()
```

```python
# =========== DENOISING ==============
```

```python
def random_signal(x_values):
    a = r.uniform(-1.5,1.5)
    b = r.uniform(0,2)
    result = np.array([b/((x-a)**2 + 1) for x in x_values])
    return result

def add_noise(signal):
    noise = np.array([r.uniform(-1,1)/10 for _ in range(20)])
    return signal+noise
vec_add_noise = np.vectorize(add_noise)
```

```python
x_values = np.linspace(-2,2, num=20)
sig = random_signal(x_values)
noisy_sig = add_noise(sig)

plt.plot(sig)
plt.plot(noisy_sig)
plt.show()
```

```python
y = np.array([random_signal(x_values) for _ in range(20)])
x = add_noise(y)
```

```python
x_in = Input(shape=(20,))
s = Dense(6, activation='sigmoid')(x_in)
s = Dense(20, activation='linear')(s)

model = Model(inputs=x_in, outputs=s)
model.summary()
```

```python
model.compile(optimizer='SGD', loss='mean_squared_error')
```

```python
model.fit(x, y, epochs=1000, batch_size=10)
```

```python
y_pred = model.predict(x)
```
